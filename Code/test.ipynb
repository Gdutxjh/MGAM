```python
n = 4
k =3
a = [3,4,6,18]
mod = int(1e9+7)

def gcd(x, y):
    if y == 0:
        return x
    else:
        return gcd(y, x % y)

def solve():
    cnt = 0
    for mask in range(1 << n):
        cur_gcd = -1
        for i in range(n):
            if mask & (1 << i):
                if cur_gcd == -1:
                    cur_gcd = a[i]
                else:
                    cur_gcd = gcd(cur_gcd, a[i])
        if cur_gcd == k:
            cnt += 1
    return cnt % mod

print(solve())
```
